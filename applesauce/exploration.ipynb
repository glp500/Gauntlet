# Teen Mental Health Dataset: Executive Overview & Relationships

Patterns in Mental Health, Demographics, Digital Behaviors, and Academic Performance

**Theme:** Boardroom Slate

## Analysis Plan

This analysis will give a high-level summary of who is in the dataset, how common various mental health concerns are among these teens, and patterns in social media use, sleep, and physical activity. Simple descriptive statistics and visualizations will highlight key distributions. Comparative and correlation analyses will shed light on possible links between lifestyle/digital behaviors and mental health markers. The goal is to provide accessible and actionable insights for further investigation or intervention planning.

- What are the demographic characteristics (age, gender) of the sample?
- How are mental health indicators (stress, anxiety, addiction, depression) distributed?
- What are the patterns for lifestyle variables (social media use, sleep, physical activity)?
- Are there notable associations between digital/social behaviors and mental health outcomes?
- How does academic performance vary across other variables?

```python
from pathlib import Path

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

DATA_PATH = Path('cleaned_data.csv')
df = pd.read_csv(DATA_PATH)

THEME = {
  "name": "executive",
  "label": "Boardroom Slate",
  "palette": [
    "#0F172A",
    "#2563EB",
    "#14B8A6",
    "#F59E0B",
    "#E2E8F0"
  ],
  "background": "#F8FAFC",
  "foreground": "#0F172A",
  "accent": "#2563EB",
  "plotly_template": "plotly_white"
}
PALETTE = THEME["palette"]

def _display_plotly_table(table_df, title):
    preview = table_df.copy()
    preview = preview.astype("object").where(pd.notna(preview), "")
    fig = go.Figure(
        data=[
            go.Table(
                header=dict(values=list(preview.columns), fill_color=THEME["accent"], font=dict(color="white")),
                cells=dict(values=[preview[column].tolist() for column in preview.columns], fill_color="white"),
            )
        ]
    )
    fig.update_layout(title=title, paper_bgcolor=THEME["background"], font=dict(color=THEME["foreground"]))
    fig.show()
```

## Sample Overview (First 5 Rows)

A preview of the dataset to give stakeholders a sense of the data structure and content.

```python
table_df = df.head(5)
_display_plotly_table(table_df, 'Sample Overview (First 5 Rows)')
```

## Summary Statistics for Numeric Variables

Descriptive statistics (mean, median, std, min, max) for all numeric columns, highlighting central tendency and dispersion in mental health and lifestyle indicators.

```python
table_df = df.describe(include="number").transpose().reset_index().rename(columns={"index": "column"})
_display_plotly_table(table_df, 'Summary Statistics for Numeric Variables')
```

## Distribution of Key Categorical Variables

Distribution (counts and percentages) for gender, platform_usage, and social_interaction_level, giving insight into major subgroups.

```python
parts = []
for column in df.select_dtypes(include=["object", "category", "bool"]).columns:
    counts = df[column].astype("string").fillna("<missing>").value_counts().head(5).reset_index()
    counts.columns = ["value", "count"]
    counts.insert(0, "column", column)
    parts.append(counts)
table_df = pd.concat(parts, ignore_index=True) if parts else pd.DataFrame({"note": ["No categorical columns detected."]})
_display_plotly_table(table_df, 'Distribution of Key Categorical Variables')
```

## Distribution of Gender and Social Media Platform Usage

Shows the demographic distribution of the sample by gender and popular social media platforms used, clarifying the cohort’s composition and baseline digital behaviors.

```python
plot_df = df['platform_usage'].astype("string").fillna("<missing>").value_counts().head(20).rename_axis('platform_usage').reset_index(name="count")
fig = px.bar(plot_df, x='platform_usage', y="count", color_discrete_sequence=PALETTE)
fig.update_layout(
    title='Distribution of Gender and Social Media Platform Usage',
    template=THEME["plotly_template"],
    paper_bgcolor=THEME["background"],
    plot_bgcolor=THEME["background"],
    font=dict(color=THEME["foreground"]),
)
fig.show()
```

## Distributions of Stress, Anxiety, and Addiction Levels

Visualize the prevalence and patterns of key mental health indicators among teens using histograms for stress_level, anxiety_level, and addiction_level.

```python
fig = px.histogram(df, x='stress_level', color_discrete_sequence=PALETTE)
fig.update_layout(
    title='Distributions of Stress, Anxiety, and Addiction Levels',
    template=THEME["plotly_template"],
    paper_bgcolor=THEME["background"],
    plot_bgcolor=THEME["background"],
    font=dict(color=THEME["foreground"]),
)
fig.show()
```

## Depression by Gender

Counts and percentages of depression_label stratified by gender; helps identify disparities in depression rates between genders.

```python
parts = []
for column in df.select_dtypes(include=["object", "category", "bool"]).columns:
    counts = df[column].astype("string").fillna("<missing>").value_counts().head(5).reset_index()
    counts.columns = ["value", "count"]
    counts.insert(0, "column", column)
    parts.append(counts)
table_df = pd.concat(parts, ignore_index=True) if parts else pd.DataFrame({"note": ["No categorical columns detected."]})
_display_plotly_table(table_df, 'Depression by Gender')
```

## Social Interaction Level by Platform Usage

Cross-tabulation of social_interaction_level against platform_usage to reveal how digital platforms relate to in-person social engagement.

```python
parts = []
for column in df.select_dtypes(include=["object", "category", "bool"]).columns:
    counts = df[column].astype("string").fillna("<missing>").value_counts().head(5).reset_index()
    counts.columns = ["value", "count"]
    counts.insert(0, "column", column)
    parts.append(counts)
table_df = pd.concat(parts, ignore_index=True) if parts else pd.DataFrame({"note": ["No categorical columns detected."]})
_display_plotly_table(table_df, 'Social Interaction Level by Platform Usage')
```

## Associations: Social Media Use & Sleep with Stress Level

Explore relationships between lifestyle behaviors (daily social media hours, sleep hours) and stress levels to flag potential risk patterns.

```python
fig = px.scatter(df, x='daily_social_media_hours', y='stress_level', color='sleep_hours', color_discrete_sequence=PALETTE)
fig.update_layout(
    title='Associations: Social Media Use & Sleep with Stress Level',
    template=THEME["plotly_template"],
    paper_bgcolor=THEME["background"],
    plot_bgcolor=THEME["background"],
    font=dict(color=THEME["foreground"]),
)
fig.show()
```

## Correlation Matrix for Numeric Variables

Matrix of pairwise correlations among numeric columns (e.g., stress_level, anxiety_level, addiction_level, sleep_hours, academic_performance) to identify potential associations.

```python
table_df = df.describe(include="number").transpose().reset_index().rename(columns={"index": "column"})
_display_plotly_table(table_df, 'Correlation Matrix for Numeric Variables')
```

## Academic Performance by Depression Status and Gender

Compare academic performance across depression label and gender to explore possible disparities.

```python
fig = px.box(df, x='depression_label', y='academic_performance', color_discrete_sequence=PALETTE)
fig.update_layout(
    title='Academic Performance by Depression Status and Gender',
    template=THEME["plotly_template"],
    paper_bgcolor=THEME["background"],
    plot_bgcolor=THEME["background"],
    font=dict(color=THEME["foreground"]),
)
fig.show()
```
